```python
import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
path_to_folder = "/home/lorenzo/Documents/my_papers/ROBOT2023/data"
path_to_all_train_loss = os.path.join(path_to_folder, "every_train_loss.json")
path_to_train_loss_pe= os.path.join(path_to_folder, "train_loss_per_epoch.json")
path_to_test_loss_pe = os.path.join(path_to_folder, "test_loss_per_epoch.json")
with open(path_to_all_train_loss, "r") as f:
    all_train_loss = json.load(f)
with open(path_to_train_loss_pe, "r") as f:
    train_loss_pe = json.load(f)
with open(path_to_test_loss_pe, "r") as f:
    test_loss_pe = json.load(f)
```

```python
all_train_loss = np.array(all_train_loss)
print(all_train_loss.shape)
test_loss_pe = np.array(test_loss_pe)
print(test_loss_pe.shape)
train_loss_pe= np.array(train_loss_pe)
print(train_loss_pe.shape)
```

```python
figure = plt.figure(figsize=(10,5))
axis = figure.add_axes([0,0,2,2])
large_x_axis = np.linspace(0,64, len(all_train_loss))
small_x_axis = np.arange(1,65)
plt.plot(large_x_axis, all_train_loss[:,2],linewidth=4)
#plt.plot(small_x_axis, train_loss_pe[:,2],linewidth=3)
plt.plot(small_x_axis, test_loss_pe[:,2],linewidth=6)
plt.ylabel('Loss', fontweight='bold',  fontsize='30', horizontalalignment='center')
plt.yscale("log")
plt.xticks(fontsize=40)
plt.yticks(fontsize=40)
plt.legend(["Train Toss", "Test Loss"], fontsize=40)
axis.set_yticks([0.1,0.05, 0.01,0.005, 0.001],[0.1,0.05, 0.01,0.005, 0.001])
plt.ylabel('MSE Loss (rad²)', fontweight='bold',  fontsize='40', horizontalalignment='center')
plt.xlabel('Epochs', fontweight='bold', fontsize='40', horizontalalignment='center')
plt.show()
```
